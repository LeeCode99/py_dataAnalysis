# cluster a column of text entries (using an appropriate clustering method)


```python
import warnings
warnings.filterwarnings('ignore')
# import numpy, AffinityPropagation, distance, pandas. 
import numpy as np
from sklearn.cluster import AffinityPropagation
import distance
import pandas as pd

# to print the dictionary neatly and in a format that is easy to visualize 
def print_inventory(dct):
    print("Item on the left represents each of the value, right in parenthesis is number of values:")
    for item, amount in dct.items():  # dct.iteritems() in Python 2
        print("{} ({})".format(item, amount))


# clustering method takes an input of column
def clustering(col):
    dict = {}        # empty dictionary
    df = pd.read_csv('Portal_rodents_19772002_scinameUUIDs.csv',encoding='windows-1254')
    # replace all the null values with empty string
    df.fillna('', inplace=True) # you can see the output that has '' with no value and counts
    
    # for loop that goes through each of the variable in a column and append to the dictionary
    for val in df[col]:
        # if the variable is inside the dictionary, it adds a count
        if val in dict:
            dict[val] = dict[val] + 1
        # if the variable is not inside the dictionary, it appends a new value and a value of 1
        else:
            dict[val] = 1
    # print out the value of dictionary by calling the function print_inventory
    print_inventory(dict)
```

```python
# trying with 'localiity' column
clustering('locality')
# as you can see there is no duplicates and it shows the count of that value
# the sum of the counts is equivalent to the number of rows in a column
```

```python
# using clustering function with a column called 'species'
clustering('species')
# as you can see there is no duplicates and it shows the count of that value
# the sum of the counts is equivalent to the number of rows in a column
```

# Trimming all the white spaces and replacing internal whitespace by a single space.

```python
# replace the whitespaces
def trimmerStr(df, col):
    # show the original values with whitespaces
    print(df[col].loc[:5]) 
    newdf = df
    # replace whitespaces
    newdf[col] = newdf[col].str.replace(" ","")
    # print after replacement of whitespaces
    print(newdf[col].loc[:5])
```

```python
# An example of scientificName column before and after
df = pd.read_csv('Portal_rodents_19772002_scinameUUIDs.csv',encoding='windows-1254')
trimmerStr(df,'scientificName')
```

# Sorting and clustering same values without having duplicates.

```python
import numpy as np
from sklearn.cluster import AffinityPropagation
import distance
# import numpy, AffinityPropagation, distance

#function to remove all the white space and the function takes an input of data frame
def removeWhiteSpace(data):
    # initizlize two empty array to contain the values from the column
    arr = []
    arr2 = []
    
    # for each variable that might have the white space goes through the for loop
    for val in data:
        # words assinged as array that contains words with no white space
        words = val.split(" ") 
        # the words now can be located as an arrary with numpy
        words = np.asarray(words) 
        
        gapCheck = -1*np.array([[distance.levenshtein(w1,w2) for w1 in words] for w2 in words])

        affprop = AffinityPropagation(affinity="precomputed", damping=0.5)
        affprop.fit(gapCheck)
        for cluster_id in np.unique(affprop.labels_):
            # group the variables that are close to each other
            exemplar = words[affprop.cluster_centers_indices_[cluster_id]]
            # checking the unique value of each variables
            cluster = np.unique(words[np.nonzero(affprop.labels_==cluster_id)])
            cluster_str = ", ".join(cluster)
            # appends to the array that has two cluster of the values and have no white spaces
            if exemplar not in arr:
                arr.append(exemplar)                
            if cluster_str not in arr2:
                arr2.append(cluster_str)

    print(arr)
```

```python
df_new = df[df['scientificName'].notnull()]
removeWhiteSpace(df_new['scientificName'])
```

```python
df_new = df[df['species'].notnull()]
removeWhiteSpace(df_new['species'])
```

```python
df_new = df[df['country'].notnull()]
removeWhiteSpace(df_new['country'])
```
